```python
# IMPORT AND CONVERT BZ2 FILE TO CSV
print(__name__)
import bz2
import csv
import json

testBz2File = bz2.open('../Dataset/sample/00.json.bz2', mode='rt');
print(testBz2File);
print(type(testBz2File));
# print(testBz2File.peek());

# csv read
# with testBz2File as csvfile:
#     reader = csv.DictReader(csvfile)
# #     print(reader.keys()[0][0])
#     i = 0;
#     for row in reader:
#         print(i)
#         i += 1
#         if i == 1:
#             print(row.keys())


# JSON read
lines = []
for i, line in enumerate(testBz2File):
    tweets = json.loads(line)
    if "delete" not in tweets.keys():
        testTweets = {k:v for k, v in tweets.items() if k in ("lang","text")}
#         we can extract hashstags directly, but we gonna modify it to not
#         include indices
#         testTweets['hashtags'] = tweets['entities']['hashtags']
        testTweets['hashtags'] = []
        for TweetHashtags in tweets['entities']['hashtags']:
            testTweets['hashtags'].append(TweetHashtags.get('text'))
        testTweets['linefrom'] = i
#         Taking only english tweets for easier prediction
        if testTweets["lang"] == "en":
            lines.append(testTweets)

# print(type(lines))
# print(lines)
```
